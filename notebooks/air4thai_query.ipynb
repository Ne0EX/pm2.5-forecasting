```python
new_data = requests.get("http://air4thai.pcd.go.th/services/getNewAQI_JSON.php")
station_list_temp = [station['stationID'] for station in dict(new_data.json())['stations']]

# Select only known sensor (Should be well written)
station_pattern = re.compile("^[0-9].*t$")
station_list = []
for station in station_list_temp:
  if station_pattern.match(station): station_list.append(station)

# Parameter setting (Only 1 month query in hourly)
param_dict = {
    "start_date": "2024-06-02",
    "start_time": "00",
    "end_date": "2024-07-02",
    "end_time": "23"
}

station_list_str = ",".join(station_list)

url3 = f"http://air4thai.com/forweb/getHistoryData.php?stationID={station_list_str}&param=PM25&type=hr&sdate={param_dict['start_date']}&edate={param_dict['end_date']}&stime={param_dict['start_time']}&etime={param_dict['start_time']}"
r3 = requests.get(url3)1
temp_data = dict(r3.json())
```

```python
# Get hourly data
data = []

for i in range(len(temp_data['stations'])):
  data.append(pd.DataFrame.from_dict(temp_data['stations'][0]['data']).set_index('DATETIMEDATA'))

data = pd.concat(data, axis=1)
data.columns = station_list
data.index.names = ['Date']
```
